# Exploratory Data Analysis - Yelp Dataset

## Import Packages

```python
import json
import pandas as pd
# import re
import matplotlib.pyplot as plt
```

## Load Data

```python
files = ['yelp_academic_dataset_business.json',
         'yelp_academic_dataset_review.json',
         'yelp_academic_dataset_checkin.json',
         'yelp_academic_dataset_tip.json',
         'yelp_academic_dataset_user.json']
yelp_dfs = {}
for file in files:
    name = file.split('.')[0].split("_")[-1]
    data_file = open(file)
    data = []
    for line in data_file:
        data.append(json.loads(line))
    df = pd.DataFrame(data)
    yelp_dfs[name] = df
    data_file.close()
```

```python
business_df = yelp_dfs['business']
review_df = yelp_dfs['review']
checkin_df = yelp_dfs['checkin']
tip_df = yelp_dfs['tip']
user_df = yelp_dfs['user']
```

## Pennsylvania Businesses

```python
#business_df.head()
pa_businesses = business_df[business_df['state'] == 'PA'] ## 34,039 businesses

```

```python
pa_businesses.head()
```

```python
pa_businesses.groupby(by="city")['state'].count().sort_values(ascending=False)
```

Business categories in PA

```python
# Extract count of all categories
pa_category_count = {}
for row in pa_businesses['categories'].dropna():
    categories = row.split(",")
    for category in categories:
        category = category.strip()
        if not category in pa_category_count:
            pa_category_count[category] = 1
        else:
            pa_category_count[category] += 1
```

```python
print(pa_category_count)
```

```python
# convert to df & take top 25
pa_category_df = pd.DataFrame.from_dict(pa_category_count, orient='index', columns=['count'])
pa_category_df = pa_category_df.sort_values(by='count',ascending=False).head(25).sort_values(by='count',ascending=True)

# plot
pa_category_df.plot(kind='barh', y='count', legend=None)
plt.title('PA Businesses Category Count')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

Join PA Business to Reviews

```python
pa_joined = pa_businesses.merge(review_df,on='business_id',how='left')
pa_joined.head()
```

```python
pa_rev = pa_joined.groupby(by=['business_id','name','stars_x','postal_code','latitude','longitude','city']).agg({'stars_y':'mean', 'useful':'sum', 'funny':'sum', 'cool':'sum', 'review_id':'count'})
pa_rev = pa_rev.reset_index().rename(columns={'review_id':'review_count'})
pa_rev['stars_y'] = round(pa_rev['stars_y'],2)
pa_rev = pa_rev.rename(columns={'stars_x':'yelp_avg_rating','stars_y':'calc_avg_rating'}).sort_values(by='review_count',ascending=False)
pa_rev.head(10)
```

```python
# Highest rated PA businesses with > 100 reviews
pa_rev[pa_rev['review_count'] > 100].sort_values(by='calc_avg_rating',ascending=False).head(10)
```

```python
# Lowest rated PA businesses with > 100 reviews
pa_rev[pa_rev['review_count'] > 100].sort_values(by='calc_avg_rating',ascending=True).head(10)
```
