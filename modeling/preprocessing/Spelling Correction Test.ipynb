And I noticed that one step we did not include in our preprocessing (Compared to "Resampling imbalanced data to detect fake reviews using machine learning classifiers and textual-based features" written by Budhi et al., in which TF-IDF and SVM are used on YelpZip) is spelling correction. This step is important because misspelled words add unnecessary difficulty to detection. If they are not removed, they may be recognized as different words from correct forms. ## How to solve the problem? Maybe we can treat interjections differently

### First, a list of interjection words was created.

```python
interjections_list=['aha','ahem','ahh','ahoy','alas','arg','aw', 'bam', 'bingo', 
          'blah', 'boo', 'bravo', 'brrr', 'cheers', 'congratulations', 
          'dang', 'drat', 'darn','duh', 'eek', 'eh', 'encore', 'eureka', 
          'fiddlesticks', 'gadzooks', 'gee', 'golly', 'goodness', 'gosh',
          'haha', 'hallelujah', 'hey', 'hmm', 'huh','hurray', 'humph', 
          'oh', 'oops', 'ouch', 'ow', 'phew', 'phooey', 'pooh', 'pow', 
          'shh', 'shoo', 'ugh', 'wahoo', 'whoa', 'whoops', 'wow', 
          'yikes', 'yeah', 'yippee', 'yo', 'yuck']
```

### Then, interjections were remained without being over corrected

```python
def correct_words3(list_words):
    corrected=[]
    for w in list_words:
        if w not in interjections_list:
            corrected.append(str(TextBlob(w).correct())) #Getting the object for the word
        else:
            corrected.append(w)

    correct_words=words.words()
    new_corrected=[]
    for word in corrected:
        if word not in interjections_list:
            if len(word)<=1: #avoid division by zero and out of indexer problem 
                new_corrected.append(word)
            else:
                temp = [(jaccard_distance(set(ngrams(word, 2)),
                                          set(ngrams(w, 2))), w)
                        for w in correct_words if w[0] == word[0]]
                new_corrected.append(sorted(temp, key=lambda val: val[0])[0][1])
        else:
            new_corrected.append(word)
    corrected2=[]
    for w in new_corrected:
        if w not in interjections_list:
            corrected2.append(str(TextBlob(w).correct())) #Getting the object for the word
        else:
            corrected2.append(w)

    return corrected2
```
